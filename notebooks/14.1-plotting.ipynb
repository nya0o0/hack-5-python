```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn.datasets import load_iris
from sklearn.linear_model import LinearRegression
```

```python
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
iris_df
```

```python
# Define a variable for the feature we wish to plot
feature = "sepal length (cm)"
# Group the data in the dataframe by species ID
gb = iris_df.groupby("species")
# Iterate through the groupby object
for sp, dat in gb:
    # For each species, plot the data for the given feature as a histogram
    plt.hist(dat[feature], label=sp)
```

```python
feature = "sepal length (cm)"
gb = iris_df.groupby("species")
for sp, dat in gb:
    plt.hist(dat[feature], label=sp, alpha=0.3)
plt.legend()
```

```python
feature = "sepal length (cm)"
gb = iris_df.groupby("species")
# Create a dictionary mapping species IDs to color names, which we will access inside the for loop
cdict = {'setosa':'cornflowerblue',
         'versicolor':'salmon',
         'virginica':'goldenrod'}
for sp, dat in gb:
    plt.hist(dat[feature], label=sp, alpha=0.5, color=cdict[sp])
```

```python
x_feature = "sepal length (cm)"
y_feature = "sepal width (cm)"

gb = iris_df.groupby("species")
cdict = {'setosa':'cornflowerblue',
         'versicolor':'salmon',
         'virginica':'goldenrod'}
for sp, dat in gb:
    plt.scatter(dat[x_feature], dat[y_feature], marker="*", label=sp, alpha=0.75, color=cdict[sp])

plt.title(f'Histogram of {x_feature} and {y_feature} by Species')
plt.ylabel(y_feature)
plt.xlabel(x_feature)
plt.legend()
plt.savefig('sepal_length & sepal_width.png')
```

```python
x_feature = "sepal length (cm)"
y_feature = "sepal width (cm)"
z_feature = "petal length (cm)"

gb = iris_df.groupby("species")
cdict = {'setosa':'cornflowerblue',
         'versicolor':'salmon',
         'virginica':'goldenrod'}
for sp, dat in gb:
    plt.scatter(dat[x_feature], dat[y_feature], s=dat[z_feature] * 20, marker="*", label=sp, alpha=0.75, color=cdict[sp])

plt.title(f'Scatter plot of {x_feature} vs {y_feature}, with marker size representing {z_feature}')
plt.xlabel(x_feature)
plt.ylabel(y_feature)
plt.legend()
plt.savefig('3D.png')
plt.show()
```

```python
import requests
```

```python
url = "https://esapubs.org/archive/ecol/E084/093/Mammal_lifehistories_v2.txt"
response = requests.get(url)

# Save the file locally
with open("Mammal_lifehistories_v2.txt", "wb") as file:
    file.write(response.content)
```

```python
# Load the dataset
df = pd.read_csv("Mammal_lifehistories_v2.txt", sep="\t", na_values=['-999', '-999.00'])

# Display basic info
print(df.head())
print(df.info())
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df["mass(g)"], df["newborn(g)"], alpha=0.75, color = "cornflowerblue")
plt.xlabel("Adult Mass (g)")
plt.ylabel("Newborn Mass (g)")
plt.title("Adult Mass vs. Newborn Mass")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(np.log10(df["mass(g)"]), np.log10(df["newborn(g)"]), alpha=0.5)
plt.xlabel("log10(Adult Mass) (g)")
plt.ylabel("log10(Newborn Mass) (g)")
plt.title("Log-Transformed Adult Mass vs. Newborn Mass")
plt.grid(True, linestyle="--", linewidth=0.5)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(np.log10(df["mass(g)"]),df["litter size"], alpha=0.5)
plt.xlabel("log10(Adult Mass (g))")
plt.ylabel("Litter Size")
plt.title("Adult Mass vs. Litter Size")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```
